```python
# ============================================
# ADVANCED NLP PREPROCESSING ENGINE
# ============================================

import re
from collections import Counter

# ============================================
# TASK 1: THEORY OUTPUT
# ============================================

print("\n========== TASK 1: THEORY ==========\n")

print("Q1: 'Love' vs 'love'")
print("-> Treated as different tokens due to case sensitivity.")
print("-> Normalization reduces vocabulary size.\n")

print("Q2: Without stopword removal")
print("-> Data becomes noisy.")
print("-> Common words dominate model learning.\n")

print("Q3: When stopword removal is harmful")
print("-> Sentiment: 'not good' loses meaning if 'not' removed.")
print("-> QA systems: removing 'where', 'why' breaks intent.\n")

print("Q4: Stemming vs Lemmatization")
print("-> Stemming: crude chopping (playing → playin)")
print("-> Lemmatization: dictionary-based (playing → play)\n")


# ============================================
# TASK 2: TEXT CLEANING FUNCTION
# ============================================

def clean_text_pipeline(raw_input):
    """
    Transform noisy input text into meaningful tokens.
    """

    try:
        # Validate input
        if type(raw_input) != str or raw_input.strip() == "":
            return []

        processed = raw_input.lower()

        processed = re.sub(r'https?://\S+|www\.\S+', '', processed)
        processed = re.sub(r'\S+@\S+', '', processed)

        processed = re.sub(r'\d+', '', processed)

        processed = re.sub(r'[^\w\s]', '', processed)

        processed = re.sub(r'(.)\1{2,}', r'\1', processed)

        processed = re.sub(r'\s+', ' ', processed).strip()

        word_list = processed.split()

        final_words = []
        for w in word_list:
            if len(w) > 2 or w in ["no", "not"]:
                final_words.append(w)

        return final_words

    except Exception as err:
        print("Processing error:", err)
        return []


# ============================================
# TASK 3: TEST CASE EXECUTION
# ============================================

sample_data = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

print("\n========== TASK 3: OUTPUT ==========\n")

processed_store = []

for txt in sample_data:
    words = clean_text_pipeline(txt)
    rebuilt = " ".join(words)

    print("Input :", txt)
    print("Tokens:", words)
    print("Clean :", rebuilt)
    print("-" * 55)

    processed_store.append((txt, words))


# ============================================
# TASK 4: STATISTICS
# ============================================

print("\n========== TASK 4: STATISTICS ==========\n")

for sent, words in processed_store:

    if words:
        count_total = len(words)
        count_unique = len(set(words))
        avg_size = round(sum(len(i) for i in words) / count_total, 2)
    else:
        count_total = count_unique = avg_size = 0

    print({
        "Text": sent,
        "Token Count": count_total,
        "Unique Count": count_unique,
        "Avg Length": avg_size
    })

print("\nInsights:")
print("-> Most noisy: 'Win $$$ now!!! Limited offer!!!'")
print("-> Most meaningful: 'I am not happy with this'")


# ============================================
# TASK 5: WORD FREQUENCY
# ============================================

print("\n========== TASK 5: WORD DISTRIBUTION ==========\n")

combined_words = []

for _, token_set in processed_store:
    combined_words += token_set

freq_map = Counter(combined_words)

print("Top 10 Words:")
print(freq_map.most_common(10))

print("\nLeast 5 Words:")
print(freq_map.most_common()[-5:])


# ============================================
# TASK 6: COMPLETE PIPELINE
# ============================================

def run_full_process(input_collection):
    master_tokens = []
    sentences_cleaned = []

    for line in input_collection:
        tokens_here = clean_text_pipeline(line)
        joined_line = " ".join(tokens_here)

        master_tokens.extend(tokens_here)
        sentences_cleaned.append(joined_line)

    return {
        "tokens": master_tokens,
        "clean_sentences": sentences_cleaned
    }


print("\n========== TASK 6: PIPELINE RESULT ==========\n")
print(run_full_process(sample_data))


# ============================================
# TASK 7: EDGE CASE TESTING
# ============================================

print("\n========== TASK 7: EDGE CASES ==========\n")

test_cases = ["", "🤣🤣🤣🤣", "999999999"]

for item in test_cases:
    print("Given :", item)
    print("Result:", clean_text_pipeline(item))
    print("-" * 50)
```
